```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)
x, y = np.meshgrid(x,y)
z = 2*x + 5*y
ax.plot_wireframe(x,y,z)
ax.plot_wireframe(2*x,2*y,2*z)
```

```python
fig = figure()
ax = Axes3D(fig)
x = np.arange(-5,5,0.5)
y = np.arange(-5,5,0.5)
x, y = np.meshgrid(x,y)
z = np.sqrt(x**2+y**2)
ax.plot_wireframe(x,y,z)
```
